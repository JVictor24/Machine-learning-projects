```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Misal kamu punya dataset heart.csv
df = pd.read_csv('heart_statlog_cleveland_hungary_final_cleaned.csv')  # pastikan file ini tersedia di folder kerja

# Fitur & Target
X = df.drop('target', axis=1)
y = df['target']

# Standarisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA ke 2 dimensi
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualisasi sebelum training
plt.figure(figsize=(6,5))
plt.scatter(X_pca[y==0, 0], X_pca[y==0, 1], color='blue', label='Tidak ada penyakit')
plt.scatter(X_pca[y==1, 0], X_pca[y==1, 1], color='red', label='Ada penyakit')
plt.title("Visualisasi Data Sebelum Training (PCA)")
plt.xlabel("Komponen Utama 1")
plt.ylabel("Komponen Utama 2")
plt.legend()
plt.grid(True)
plt.show()

# Train Logistic Regression
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
model = LogisticRegression()
model.fit(X_train, y_train)

# Prediksi di data PCA (untuk boundary)
def plot_decision_boundary(model, X_pca, y):
    x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
    y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                         np.linspace(y_min, y_max, 300))
    
    # Karena model dilatih di space asli, kita transform grid ke original space
    grid = np.c_[xx.ravel(), yy.ravel()]
    grid_original_space = pca.inverse_transform(grid)
    
    Z = model.predict(grid_original_space)
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(6,5))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.RdBu)
    plt.scatter(X_pca[y==0, 0], X_pca[y==0, 1], color='blue', label='Tidak ada penyakit')
    plt.scatter(X_pca[y==1, 0], X_pca[y==1, 1], color='red', label='Ada penyakit')
    plt.title("Setelah Training Logistic Regression (PCA + Decision Boundary)")
    plt.xlabel("Komponen Utama 1")
    plt.ylabel("Komponen Utama 2")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_decision_boundary(model, X_pca, y)
```
