### Data Preprocessing dan CleanUp

#### Bagian 1: 
Pembersihan data duplikat yang dapat untuk menyebabkan bias di dalam machine learning 


```python
duplikat = df[df.duplicated()]
print("Jumlah baris duplikat:", len(duplikat))

# Hapus duplikat
df = df.drop_duplicates()
print('Done')
```

#### Bagian 2:
Dikarenakan jumlah baris dengan cholestrol 0 cukup besar yang ada 172 data (17% dari data yang sudah non duplikat), lebih baik diisikan dengan nilai median agar tidak menjadi bias di dalam machine learning

```python
chol_zeros = df[df['cholesterol'] == 0]
bp_zeros = df[df['resting bp s'] == 0]
mh_zeros = df[df['max heart rate'] == 0]

print("Jumlah baris dengan cholesterol = 0:", len(chol_zeros))
print("Jumlah baris dengan resting bp s = 0:", len(bp_zeros))
print("Jumlah baris dengan max heart rate = 0:", len(bp_zeros))

# Cleaning
df['cholesterol'] = df['cholesterol'].replace(0, df['cholesterol'].median())
df = df[df['resting bp s'] != 0]
df = df[df['max heart rate'] != 0]
```

#### Bagian 4: 
Menggunakan One Hot Encoding dari library panda untuk melakukan encoding terhadap data kategorikal seperti kolom chest pain type dan ST slope

```python
df = pd.get_dummies(df, columns=['chest pain type', 'ST slope'], drop_first=True)

print(df)
```

#### Bagian 3:
Scaling menjadi hal yang penting karena dengan scaling, mean akan dipaksa 0 dan standar deviasi menjadi 1. Hal ini akan sangat berpengaruh ke dalam pelatihan model, karena kalau ada beberapa algoritma yang akan membuat modelnya terlalu berat sebelah terhadap fitur itu. Dengan menggunakan standard scaller, beberapa pelatihan model akan lebih cepat dan stabil ketika datanya sudah distandarisasi. 

Di sisi lain, tidak ada model yang datanya tidak boleh di scalling, hal ini membuatnya tidak ada salahnya untuk melakukan standard scaler. 

```python
scaler = StandardScaler()
df[['age', 'resting bp s', 'cholesterol', 'max heart rate', 'oldpeak']] = scaler.fit_transform(
    df[['age', 'resting bp s', 'cholesterol', 'max heart rate', 'oldpeak']]
)
```

```python
Q1 = df['cholesterol'].quantile(0.25)
Q3 = df['cholesterol'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['cholesterol'] < Q1 - 1.5 * IQR) | (df['cholesterol'] > Q3 + 1.5 * IQR)]
print(outliers)

#masi mikir mau diapain 
```
